Write a Python script that reads an image from a file as grayscale, and finds the four non-overlapping 5x5 patches with highest average brightness. Take the patch centers as corners of a quadrilateral, calculate its area in pixels, and draw the quadrilateral in red into the image and save it in PNG format. Use the opencv-python package for image handling

# Steps

1. Read the image as a grayscale image 
2. Define a 5x5 patch that slides through the image.
3. Calculate the average brightness for each patch and store it by centralizing it in each dimension
4. Sort the patches based on descending order of avg brightness values
5. Get the top 4 patches and their center coordinates
6. Compute the area from patch center coordinates
7. Draw quadrilateral in red (0, 0, 255)
8. Save it as a png


# Approach 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_processing_task(image_path, num_patches = 4, patch_size = 5): #four non-overlapping patches
    # Read the image as a grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # image dimensions
    height, width = image.shape

    # Initialize a list to store patch information ((x_center,y_center), avg_brightness) to sort based on brightness value later
    patch_list = []

    # Slide through the image and calculate average brightness for each patch
    for i in range(0, height - patch_size + 1, patch_size):
        for j in range(0, width - patch_size + 1, patch_size):
            patch = image[i:i+patch_size, j:j+patch_size]
            avg_brightness = np.mean(patch)
            patch_list.append(((i + patch_size // 2, j + patch_size // 2), avg_brightness))  


    # Sort patches based on average brightness (in descending order)
    sorted_patches = sorted(patch_list, key=lambda x: x[1], reverse=True)
    
    # Take the top patches
    top_patches_list = sorted_patches[:num_patches]
    patch_centers = [center for center, _ in top_patches_list]
    
    # Calculate the area of the quadrilateral formed by patch centers
    area = cv2.contourArea(np.array(patch_centers))

    # Draw the quadrilateral in red
    quad = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    cv2.polylines(quad, [np.array(patch_centers)], isClosed=True, color=(0, 0, 255), thickness=2)

    # Save the image with the drawn quadrilateral
    cv2.imwrite('output_image.png', quad)

    return area

```

```python
#Replace image_path with your test png image
image_path = 'monalisa.png' 
area = image_processing_task(image_path) 
print(f"Area of the quadrilateral: {area} pixels")
```

```python
import cv2
import numpy as np

def add_noise_and_process(image_path, noise_type='gaussian', intensity=50):

    # Read the image
    original_image = cv2.imread(image_path)

    # Convert the image to float for some noise types
    original_image = original_image.astype(np.float32) / 255.0

    # Get image dimensions
    height, width, channels = original_image.shape

    # Generate noise based on the selected type
    noise = np.random.normal(0, intensity/255.0, (height, width, channels))

    # Add noise to the original image
    noisy_image = original_image + noise

    # Clip values to be in the valid range [0, 1]
    noisy_image = np.clip(noisy_image, 0, 1)

    # Convert back to uint8 format for displaying
    noisy_image = (noisy_image * 255).astype(np.uint8)
    
    cv2.imwrite('noisy_img.png', noisy_image)
    
    area = image_processing_task(image_path)
    
    return noisy_image

```

```python
def get_top_patches(image_path, patch_size=(5, 5), num_patches=4):
    
    # Read the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Get image dimensions
    height, width = image.shape

    # Initialize variables to store top patches and patch centers
    top_patches = []
    patch_centers = []

            
    for _ in range(num_patches):
        
        # Generate random coordinates for the top-left corner of the patch
        top_left_x = np.random.randint(0, width - patch_size[1] + 1)
        top_left_y = np.random.randint(0, height - patch_size[0] + 1)

        # Calculate the center of the patch
        center_x = top_left_x + patch_size[1] // 2
        center_y = top_left_y + patch_size[0] // 2

        # Append the center to the list
        patch_centers.append((center_x, center_y))

        # Extract the patch from the image
        patch = image[top_left_y:top_left_y + patch_size[0], top_left_x:top_left_x + patch_size[1]]

        # Calculate the average brightness of the patch
        average_brightness = np.mean(patch)

        # Append the patch and its average brightness to the list
        top_patches.append((patch, average_brightness))

    # Sort patches based on average brightness in descending order
    top_patches.sort(key=lambda x: x[1], reverse=True)

    return top_patches, patch_centers

def draw_quadrilateral(image, corners):
    # Convert the corners to integer coordinates
    corners = np.array(corners, dtype=np.int32)

    # Reshape the array for drawing
    corners = corners.reshape((-1, 1, 2))

    # Draw the quadrilateral in red
    cv2.polylines(image, [corners], isClosed=True, color=(0, 0, 255), thickness=2)

def save_image(image, output_path):
    # Save the image in PNG format
    cv2.imwrite(output_path, image)

# Example usage
image_path = 'vangogh.png'
output_path = 'output_image2.png'

top_patches, patch_centers = get_top_patches(image_path)
area = cv2.contourArea(np.array(patch_centers))
print(area)
# Create a copy of the original image to draw on
image_with_quadrilateral = cv2.imread(image_path)

# Draw the quadrilateral on the image
draw_quadrilateral(image_with_quadrilateral, patch_centers)

# Save the image with the drawn quadrilateral
save_image(image_with_quadrilateral, output_path)

print(f"Image with quadrilateral saved at: {output_path}")
```

```python
def get_top_non_overlapping_patches(image_path, patch_size=(5, 5), num_patches=4):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    height, width = image.shape
    top_patches = []
    patch_centers = []

    # Create a mask to keep track of used regions
    used_mask = np.zeros((height, width), dtype=np.uint8)

    for _ in range(num_patches):
        # Find the maximum brightness in the unused region
        max_brightness, max_brightness_loc = 0, (0, 0)

        for i in range(height - patch_size[0] + 1):
            for j in range(width - patch_size[1] + 1):
                if used_mask[i:i+patch_size[0], j:j+patch_size[1]].sum() == 0:
                    patch = image[i:i+patch_size[0], j:j+patch_size[1]]
                    average_brightness = np.mean(patch)

                    if average_brightness > max_brightness:
                        max_brightness = average_brightness
                        max_brightness_loc = (i, j)

        # Mark the used region
        i, j = max_brightness_loc
        used_mask[i:i+patch_size[0], j:j+patch_size[1]] = 1

        # Calculate the center coordinates
        center_x = j + patch_size[1] // 2
        center_y = i + patch_size[0] // 2
        patch_centers.append((center_x, center_y))

        # Extract the patch
        patch = image[i:i+patch_size[0], j:j+patch_size[1]]
        top_patches.append((patch, max_brightness))

    top_patches.sort(key=lambda x: x[1], reverse=True)

    return top_patches, patch_centers

def draw_quadrilateral(image, corners):
    # Convert the corners to integer coordinates
    corners = np.array(corners, dtype=np.int32)

    # Reshape the array for drawing
    corners = corners.reshape((-1, 1, 2))

    # Draw the quadrilateral in red
    cv2.polylines(image, [corners], isClosed=True, color=(0, 0, 255), thickness=2)

def save_image(image, output_path):
    # Save the image in PNG format
    cv2.imwrite(output_path, image)

# Example usage
image_path = 'monalisa.png'
output_path = 'output_image2.png'

top_patches, patch_centers = get_top_non_overlapping_patches(image_path)
print(patch_centers)
area = cv2.contourArea(np.array(patch_centers))
print(area)
# Create a copy of the original image to draw on
image_with_quadrilateral = cv2.imread(image_path)

# Draw the quadrilateral on the image
draw_quadrilateral(image_with_quadrilateral, patch_centers)

# Save the image with the drawn quadrilateral
save_image(image_with_quadrilateral, output_path)

print(f"Image with quadrilateral saved at: {output_path}")
```

Write a Python script that reads an image from a file as grayscale, and finds the four non-overlapping 5x5 patches with highest average brightness. Take the patch centers as corners of a quadrilateral, calculate its area in pixels, and draw the quadrilateral in red into the image and save it in PNG format. Use the opencv-python package for image handling

# Steps
1. Read the image as a grayscale image 
2. Define a 5x5 patch that slides through the image.
3. Calculate the average brightness for each patch and store it by centralizing it in each dimension
4. Sort the patches based on descending order of avg brightness values
5. Get the top 4 patches and their center coordinates
6. Compute the area from patch center coordinates
7. Draw quadrilateral in red (0, 0, 255)
8. Save it as a png

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def get_top_patches(image_path, patch_size = 5, num_patches = 4, stride = 5):
    
    image = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
#     image = cv2.imread(image_path)
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#     height, width, _ = image.shape

    height, width = image.shape
    
    brightest = [] #Stores the brightest patches based on their average by sliding window 
    for i in range(0, height - patch_size + 1, stride):
        for j in range(0, width - patch_size + 1, stride):
            patch = image[i:i + patch_size, j:j + patch_size]
#             patch = gray_image[i:i + patch_size, j:j + patch_size]
            avg_brightness = np.mean(patch)
            brightest.append(((i, j), avg_brightness)) #Append the coordinates and the avg brightness
    
     #Sort the patches by average brightness in descending order
    brightest.sort(key=lambda x: x[1], reverse=True) 
    brightest = brightest[:num_patches] #get the top 4 
    
    return brightest

def construct_quad(image, brightest_patches, patch_size=5):
    
    # Coordinates and their brightness values
    for idx, ((x, y), avg_brightness) in enumerate(brightest_patches, 1):
        print(f"Patch {idx}: Position ({x}, {y}), Brightness: {avg_brightness}")
        # Visualize the patches
        cv2.rectangle(image, (y, x), (y + patch_size, x + patch_size), (0, 0, 255), 1)
        
    points = []
    for ((x, y), _) in brightest_patches:
        center_x, center_y = x + patch_size // 2, y + patch_size // 2
        points.append((center_y, center_x))  # Swap x and y for OpenCV coordinates

    points = np.array(points)
    area = cv2.contourArea(points)

    points = points.reshape((-1, 1, 2))
    
    
    cv2.polylines(image, [points], isClosed=True, color=(0, 0, 255), thickness=2)
    
    return area
    
def save_image(image, output_path):
    # Save the image in PNG format
    cv2.imwrite(output_path, image)

# Example usage
image_path = 'doggo.png'
output_path = 'output_image2.png'

brightest_patches = get_top_patches(image_path)

image = cv2.imread(image_path) 
area = construct_quad(image, brightest_patches)

save_image(image, output_path)
print(f"Area of quadrilateral in pixels: {area}")
print(f"Image with quadrilateral saved at: {output_path}")
```
